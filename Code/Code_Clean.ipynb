```python
import pandas as pd
import numpy as np
df = pd.read_csv("Sexual-Harassment-Classification/labelled_tweets.csv")
df.head(10)
```

```python
c= 0
for i in df.index:
    text = df.at[i,'Text']
    text = " ".join(filter(lambda x:x[0]!='@', text.split()))
    text = ' '.join(x for x in text.split() if not x.startswith('RT'))
    text = ' '.join(x for x in text.split() if not x.startswith('https'))
    
    df.at[i,'Text'] = text
```

```python
df = df.iloc[np.random.permutation(len(df))]
df.head()
```
